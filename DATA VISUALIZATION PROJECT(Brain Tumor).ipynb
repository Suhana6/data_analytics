```python
df=df.rename({'Size (cm)':'size'},axis=1)
df
```

```python
avg=df.groupby('Gender')['Size'].mean()
print(avg)
avg.plot(kind='bar')
plt.xlabel('gender (female=0 ,male=1)')
plt.ylabel('avg Tumor size')
plt.title('avg Tumor size by Gender')
plt.show()
```

```python
plt.hist(df['Age'],edgecolor='k')
plt.title('age distribution')
plt.show()
```

```python
plt.hist([df['Tumor Type'],df['Size']],label=['types','tumor Size (cm)'])
plt.xlabel('Tumor Type')
plt.ylabel('tumor Size')
plt.title('distribution the tumor types and size')
plt.legend()
plt.show()
```

```python
dc=df['Tumor Type'].value_counts()
print(dc)
plt.figure(figsize=(5,5))
plt.pie(dc,labels=['Astrocytoma','Ependymoma','Glioblastoma','Meningioma','Oligodendroglioma'],autopct='%1.1f%%',startangle=140,colors=['lightblue','lightyellow'])
plt.title('proportion of individuals with Tumor Type')
plt.show()
```

```python
gc=df['Gender'].value_counts()
print(gc)
plt.pie(gc,labels=['female','males'],autopct='%1.1f%%',startangle=140,wedgeprops={'width':0.6}) 
plt.title('gender distibution')
plt.show()
```

```python
bp=df['Grade'].value_counts()
print(bp)
plt.pie(bp,labels=['A','B','C','D'],autopct='%1.1f%%',startangle=140)
plt.title('tumor grade distibution')
plt.show()
```
